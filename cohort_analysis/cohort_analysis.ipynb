## Описание проекта

Необходимо помочь маркетологам оптимизировать маркетинговые затраты.
Выполнить анализ источников трафика Яндекс.Афиши за период. Определить невыгодные источники трафика, предложить рекомендации по дальнейшему распределению бюджета. Проанализировать как люди пользуются продуктом, когда начинают покупать, доходы компании по каждому клиенту и сроки окупаемости расходов на привлечение клиента в целом и в разрезе каждого источника трафика.

В распоряжении есть данные от Яндекс.Афиши с июня 2017 по конец мая 2018 года:
   - лог сервера с данными о посещениях сайта Яндекс.Афиши,
   - выгрузка всех заказов за этот период,
   - статистика рекламных расходов.

    
### Описание данных

Таблица visits (лог сервера с информацией о посещениях сайта):
   - Uid — уникальный идентификатор пользователя
   - Device — категория устройства пользователя
   - Start Ts — дата и время начала сессии
   - End Ts — дата и время окончания сессии
   - Source Id — идентификатор рекламного источника, из которого пришел пользователь

Таблица orders (информация о заказах):
   - Uid — уникальный id пользователя, который сделал заказ
   - Buy Ts — дата и время заказа
   - Revenue — выручка Яндекс.Афиши с этого заказа

Таблица costs (информация о затратах на маркетинг):
   - source_id — идентификатор рекламного источника
   - dt — дата
   - costs — затраты на этот рекламный источник в этот день

### Длительность сессии пользователя

В таблице с информацией о посещений сайта 'visits' в колонке 'duration' найдем разность между концом 'end_ts' и началом сессии 'start_ts'. Выразим результат в секундах с помощью метода dt.seconds.

```python
visits['duration'] = (visits['end_ts'] - visits['start_ts']).dt.seconds

# посмотрим первые строки таблицы
visits.head(10)
```

Среди значений длительности сессии встречаются данные, равные 0. Нулевые сессии нельзя считать результативными. Найдем их количество по дням и количество уникальных пользователй с нулевыми сессиями.

```python
visits[visits['duration']==0].groupby('day').agg({'uid':['count', 'nunique']}).head(10)
```

Как видно из первых данных количество "нулевых" сессий и количество пользователй в день с нулевой сессией почти во всех случаях равны. Это говорит о том, что при первом заходе, вероятно, произошел технический сбой и пользователь зашел на сайт повторно. При анализе средней продолжительности сессии нулевые учитывать не будем.

```python
# найдем среднюю продолжительность сессии в день без учета нулевых
mean_duration = (visits[visits['duration']>0]
                         .groupby('day')['duration'].agg('mean')
                         .reset_index()
                        )
mean_duration.head()
```

```python
# построим гистограмму распределения по продолжительности сессии 'duration'
mean_duration.hist(bins=100, figsize=(10,4))
plt.show()
```

Построим boxplot, чтобы определить выбросы

```python
mean_duration.boxplot()
plt.show()
```

```python
# увеличим масштаб
mean_duration.boxplot()
plt.ylim(500,950)
plt.show()
```

### Коэффициент удержания пользователей

Для анализа частоты возвращения людей на сервис Яндекс.Афиша будем рассчитывать коэффициент удержания Retention Rate, показывающий сколько пользователей из когорты вернулось относительно их изначального числа.

Когорты целесообразно взять длиною в месяц, так как для посещения кино, театров, концертов пользователи могут пользоваться сервисом несколько раз в месяц или реже.

```python
# найдем дату, когда пользователь впервые проявил активность
first_month = visits.groupby('uid')['month'].min().reset_index()

# переименуем колонки
first_month.columns = ['uid', 'first_month']

# объединим полученные данные с исходной таблицей
visits = visits.merge(first_month, on='uid')

visits.head()
```

```python
# определим "период жизни пользователя" 'cohort_lifetime'
visits['cohort_lifetime'] = visits['month'] - visits['first_month']

# приведем результат к порядковому номеру месяца от "начала жизни"
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')

visits.head()
```

Сгруппируем данные по устройству, первому месяцу когорты 'first_month' и сроку жизни 'cohort_lifetime'

Посчитаем для каждой когорты количество активных пользователей на каждый месяц после старта.

```python
visits_device = (visits
                 .groupby(['device','first_month', 'cohort_lifetime'])
                 .agg({'uid': 'nunique'})
                 .reset_index()
          )
visits_device.head()
```

```python
# переименуем колонки
visits_device = visits_device.rename(columns={'uid':'n_users'})
visits_device.head()
```

Чтобы найти Retention Rate, надо каждому "месяцу жизни" поставить в соответствие стартовое количество пользователей в когорте, т.е. "нулевой месяц жизни"

```python
# сформируем таблицу 'start_lifetime' из пользователй первого месяца когорты
start_lifetime = visits_device[visits_device['cohort_lifetime'] == 0][['device','first_month', 'n_users']]

# переименуем столбец 'uid' в 'start_users'
start_lifetime = start_lifetime.rename(columns = {'n_users':'start_users'})

start_lifetime

# объединим данные по когортам с исходным количеством пользователей в когорте:
visits_device = visits_device.merge(start_lifetime, on=['device','first_month'])

visits_device.head()
```

```python
# найдем Retention Rate
visits_device['retention'] = (visits_device['n_users'] / visits_device['start_users'])
visits_device.head()
```

```python
visits_device['first_month'] = visits_device['first_month'].dt.strftime('%Y-%m')
```

### Сумма валовой прибыли клиента за период (LTV)

Общий показатель LTV будем рассчитывать по месячным когортам, которые уже определены в таблице **orders_month**.

```python
# выведем первые строки таблицы
orders_month.head()
```

Для расчета общего LTV по каждой когорте необходимо валовую прибыль месяца жизни когорты разделить на общее количество пользователей в когорте. 

Количество пользователей уже рассчитано ранее. Данные в колонке 'n_buyers'. 

Необходимо определиться с валовой прибылью. Ее можно найти, зная маржинальность заказов или сумму прямых затрат. Так как этих данных нет, примем условно, что маржинальность равна 1.

```python
# считаем маржинальность равную 1
margin_rate = 1

# найдем валовую прибыль и запишем ее в колонку 'gp'
orders_month['gp'] = orders_month['revenue'] * margin_rate

orders_month['ltv'] = (orders_month['gp'] / orders_month['n_buyers']).round(2)
orders_month.head()
```

```python
# сформируем сводную таблицу для LTV по когортам и месяцам жизни
ltv_pivot = (orders_month
            .pivot_table(index='first_buy_month', 
                         columns='cohort_lifetime', 
                         values='ltv', 
                         aggfunc='sum').cumsum(axis=1).round(2)
            )

ltv_pivot.fillna('')
```

```python
# построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(12, 8))
sns.heatmap(ltv_pivot, annot=True, fmt='.3g', 
            linewidths=0.5, linecolor='gray', cmap='inferno')
plt.title('Накопительный LTV по когортам и месяцам жизни', fontsize=16)
plt.show()
```

Из всех когорт выделяется когорта '2017-09-01'. Валовая прибыль от нее нарастает быстрее всего.

Чтобы определить сколько средний пользователь генерирует LTV, возьмем период жизни 6 месцев. Этот период жизни полностью прошли когорты с '2017-06-01' по '2017-12-01'.

Рассмотрим данные таблицы ltv_pivot в выбранных границах по  'first_buy_month' и 'cohort_lifetime' c накопительным итогом:

```python
ltv_pivot.loc[:'2017-12-01', :5]
```

Рассчитаем теперь LTV в разбивке по источникам.

```python
# выведем первые строки таблицы
orders.head()
```

Для расчета показателя LTV по источникам рекламы в таблицу 'orders' необходимо включить данные по источнику рекламы.

Сформируем таблицу пользователей и источников рекламы 'user_source' из исходных данных посещений пользователй 'visits'. Источником привлечения пользователя будем считать первый источник, по которому он зашел на ресурс. Таблицу 'user_source' получим группировкой по пользователю 'uid' из таблицы посещений 'visits', предварительно отсортировав по возрастанию заходы на сайт по колонке'start_ts'

```python
# сортируем по 'start_ts', группируем по 'uid', агрегируем 'source_id' методом .first()
user_source = (visits
               .sort_values(by='start_ts')
               .groupby('uid')['source_id'].first()
               .reset_index()
              )
# посмотрим результат:
user_source.head()
```

```python
# добавим данные с источником рекламы в таблицу 'orders' по полю 'uid'
orders_source = pd.merge(orders, user_source, on='uid', how='left')

# посмотрим результат
orders_source.head()
```

Найдем число новых покупателей по источникам и когортам из таблицы 'orders' с помощью группировки по источнику рекламы и первому месяцу покупки:

```python
# в таблицу 'source_cohort_size' запишем количество уникальных пользователй 'uid' методом 'nunique' 
# c группировкой по 'first_buy_month' и 'source_id'
source_cohort_size = (orders_source
                      .groupby(['source_id', 'first_buy_month'])['uid']
                      .nunique()
                      .reset_index()
                     )
# переименуем колонки
source_cohort_size.columns = ['source_id', 'first_buy_month', 'n_buyers']
source_cohort_size.head()
```

Сгруппируем таблицу заказов 'orders_source' по источнику рекламы, месяцу первой покупки, месяцу каждого заказа и сложим выручку. Сбросим индекс методом reset_index():

Из полученных данных сформируем таблицу 'cohorts_source'

```python
# выполним групировку по столбцам 'source_id', 'first_buy_month', 'buy_month', 
# найдем количество уникальных пользователй 'nunique' и сумму доходов 'sum'
cohorts_source = (orders_source
                         .groupby(['source_id', 'first_buy_month', 'buy_month'])
                         .agg({'revenue':'sum'})
                         .reset_index()
                        )
cohorts_source.head()
```

Добавим в таблицу 'cohorts_source' данные о количестве пользователей в первый раз совершивших покупку в месяц когорты:

```python
cohorts_source = (cohorts_source
                  .merge(source_cohort_size, on=['source_id','first_buy_month'], how='left')
                        )
cohorts_source.head()
```

Найдем lifetime когорты. Из месяца совершения покупки вычтем месяц когорты:

```python
cohorts_source['cohort_lifetime'] = cohorts_source['buy_month'] - cohorts_source['first_buy_month']

cohorts_source['cohort_lifetime'] = cohorts_source['cohort_lifetime'] / np.timedelta64(1,'M')

cohorts_source['cohort_lifetime'] = cohorts_source['cohort_lifetime'].round().astype('int')

cohorts_source.head()
```

Валовую прибыль источникам и по месяцу жизни найдем, считая ранее установленную маржинальность.

```python
# найдем валовую прибыль и запишем ее в колонку 'gp'
cohorts_source['gp'] = cohorts_source['revenue'] * margin_rate
cohorts_source.head()
```

Найдем LTV по источникам. Разделим валовую прибыль когорт по источнику за каждый месяц на общее число пользователей в каждой когорте по источнику. 

```python
cohorts_source['ltv'] = (cohorts_source['gp'] / cohorts_source['n_buyers']).round(2)
cohorts_source.head()
```

Посмотрим как распределились средние доходы от пользователя по когортам в зависимости от источника рекламы.

### Окупаемость маркетинговых расходов (ROMI)

Найдем окупаемость ROMI в целом по когортам.

Воспользуемся таблицами:
   - orders_month 
   - costs 
   
Выведем таблицы для просмотра:   

```python
orders_month.head()
```

```python
costs.head()
```

Найдем САС по когортам.

```python
# сгруппируем данные таблицы costs по месяцам и найдем сумму затрат по месяцам.
costs_month = costs.groupby('dt_month')['costs'].sum()
costs_month.head()
```

```python
# присоединим к таблице orders_month данные таблицы costs_month
total_report = (orders_month[['first_buy_month','cohort_lifetime','n_buyers','gp','ltv']]
                .join(costs_month, on='first_buy_month')
               )
total_report.head()
```

```python
# найдем САС в каждой когорте
total_report['cac'] = (total_report['costs']/total_report['n_buyers']).round(2)
total_report.head()
```

```python
# найдем общий ROMI: разделим доходы одного пользователя 'ltv' на затраты одного пользователя 'cac'
total_report['romi'] = (total_report['ltv']/total_report['cac']).round(2)
total_report.head()
```

```python
# сформируем сводную таблицу
total_romi_pivot = (total_report
                    .pivot_table(index='first_buy_month',
                                 columns='cohort_lifetime',
                                 values='romi',
                                 aggfunc='mean').cumsum(axis=1)
                   )
total_romi_pivot.fillna('')                    
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(total_romi_pivot, annot=True, fmt='.3g', linewidths=0.5, linecolor='gray')
plt.title('ROMI за 12 месяцев', fontsize=14)
plt.show()
```

Лучший результат показала когорта '2017-09-01' - окупаемость на 4 месяце

Когорта '2017-06-01' - окупилась на 7 месяце жизни.


Найдем, чему равна окупаемость за 6 месяцев жизни. 
Этот период прошли когорты с '2017-06-01' по '2017-12-01'

```python
total_romi_pivot.loc[:'2017-12-01',:5]
```

```python
# найдем ROMI средней когорты за 6 месяцев
total_romi_pivot.loc[:'2017-12-01',:5][5].mean().round(2)
```

Итак, за 6 месяцев жизни средняя когорта окупается только на **86%**

Найдем ROMI по источникам рекламы. Воспользуемся данными таблицы **report**. Разделим доходы от одного пользователя 'ltv' на затраты на одного пользователя 'cac':

```python
report['romi'] = (report['ltv'] / report['cac']).round(2)
report.head()
```

Построим сводные таблицы и графики ROMI по источникам рекламы.

```python
# создадим словарь источников рекламы для последующего обращения в цикле
source_dict = {'1':'источник 1', '2':'источник 2', '3':'источник 3', '4':'источник 4',
               '5':'источник 5', '9':'источник 9', '10':'источник 10'}
```

```python
# обратимся в цикле к каждому источнику рекламы по ключасм словаря source_dict
# сформируем для кажого источника сводную таблицу по когортам и их возрастам 
# с накопительным значением ROMI 
# построим визуализацию 'heatmap'

for source in source_dict.keys():
    romi_pivot = (report.query('source_id == @source')
                    .pivot_table(index='first_buy_month', 
                                 columns='cohort_lifetime', 
                                 values='romi', 
                                 aggfunc='mean').cumsum(axis=1).round(2)
                   )
    sns.set(style='white')
    plt.figure(figsize=(12, 8))
    sns.heatmap(romi_pivot, annot=True, fmt='.3g', 
                linewidths=0.5, linecolor='gray', cmap='inferno')
    plt.title('ROMI, {}'.format(source_dict[source]), fontsize=14)
    plt.show()
```

Окупаемость по источникам:

    1  - окупаемость самая эффективная до когорты 2017-11 (первый и второй месяц жизни), затем снизилась
 
    2 - окупаемость очень различается  по когортам:
      - 2017-06 - 7-й месяц
      - 2017-07 - 11-й месяц
      - 2017-09 - 2-й месяц
      - 2017-10 - 7-й месяц
      - когорты 2017-08, 2017-11 окупаются очень медлено
     
    3 - не окупился за весь период
 
    4 - только две когорты окупились
      - 2017-06 на 7-й месяц
      - 2017-11 на первом
      - остальные не окупились
      
    5 - окупились только 3 когорты на 3-4 месяцы
    
    9 - окупились первые 5 когорт

    10 - окупились 2 когорты
